```python
import requests
from bs4 import BeautifulSoup

def scrape_job_listings(url, num_pages):
    job_listings = []

    for page in range(1, num_pages + 1):
        page_url = f"{url}&start={page * 25}"

        response = requests.get(page_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        listings = soup.find_all('li', class_='result-card')

        for listing in listings:
            job_title = listing.find('h3', class_='result-card__title').text.strip()
            company_name = listing.find('h4', class_='result-card__subtitle').text.strip()
            location = listing.find('span', class_='job-result-card__location').text.strip()

            job_listings.append({
                'Job Title': job_title,
                'Company': company_name,
                'Location': location
            })

    return job_listings

# Example usage
url = "https://www.linkedin.com/jobs/search/?currentJobId=3643115590&distance=25&f_WT=2%2C1%2C3&geoId=102887311&keywords=data%20analyst"
num_pages = 20

job_listings = scrape_job_listings(url, num_pages)

# Print the scraped job listings
for job in job_listings:
    print(job)
```
